```python
import pandas as pd

df = pd.read_csv("world_top_restaurants_dataset.csv")
print(df.head())

columns = df.columns
print(columns)
```

```python
star_categry_unique = df['Country'].nunique()
star_categry_unique
```

```python
# Visualization

import matplotlib.pyplot as plt

x=df['Country']
y=df['Average_Price_USD']

plt.figure(figsize=(15, 9))
plt.xlabel("Country")
plt.ylabel("Average Price (USD)")
plt.bar(x,y)
plt.title("Average Price by country")
plt.show
```

```python
plt.figure(figsize=(15,9))
plt.bar(df["Cuisine_Type"], df["Average_Price_USD"])
plt.xlabel("Cuisine Type")
plt.ylabel("Average Price (USD)")
plt.title("Average Price by Cuisine Type")
plt.xticks(rotation=80)
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(6,4))
sns.countplot(x='Michelin_Stars', data=df)
plt.title('Count of Restaurants by Michelin Stars')
plt.xlabel('Michelin Stars')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Rating'], kde=True, bins=20)
plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Price_Category', y='Average_Price_USD', data=df)
plt.title('Box Plot of Average Price by Price Category')
plt.xlabel('Price Category')
plt.ylabel('Average Price (USD)')
plt.show()

# Bar Plot for counting restaurants by Continent
plt.figure(figsize=(8,4))
continent_counts = df['Continent'].value_counts().reset_index()
continent_counts.columns = ['Continent', 'Count']
sns.barplot(x='Continent', y='Count', data=continent_counts)
plt.title('Number of Restaurants by Continent')
plt.xlabel('Continent')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax = sns.histplot(df['Average_Price_USD'].dropna(), bins=30)
ax.set_title('Distribution of Average Price (USD)')
plt.tight_layout()
plt.show()
```

```python
#Correlation between rating and numeric features
df[['Rating','Average_Price_USD','Seating_Capacity','Years_Operating','Michelin_Stars']].corr()
```

```python
#Rating Distribution by Price Category
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df,
    x='Price_Category',
    y='Rating'
  )
plt.title('Rating Distribution by Price Category', fontsize=16) 
plt.xlabel('Price Category', fontsize=12) 
plt.ylabel('Rating', fontsize=12) 
# plt.grid(axis='y', linestyle='--', alpha=0.7) 
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
num_cols = ['Michelin_Stars','Average_Price_USD','Seating_Capacity','Rating','Years_Operating','Established_Year']
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt='.2f', vmin=-1, vmax=1, square=True)
plt.title('Correlation Matrix — numeric feature relationships')
plt.show()
```
